```python
house['lease_type'] = house['lease_type'].map({'FAMILY':3,'ANYONE':1,'BACHELOR':0,'COMPANY':2})
house['furnishing'] = house['furnishing'].map({'SEMI_FURNISHED':1,'NOT_FURNISHED':0,'FULLY_FURNISHED':2})
house['parking'] = house['parking'].map({'BOTH':3,'TWO_WHEELER':1,'FOUR_WHEELER':2,'NONE':0})
house['water_supply'] = house['water_supply'].map({'CORP_BORE':1,'CORPORATION':0,'BOREWELL':2})
```

```python
house['internet'] = house['internet'].astype(int)
house['ac'] = house['ac'].astype(int)
house['club'] = house['club'].astype(int)
house['intercom'] = house['intercom'].astype(int)
house['servant'] = house['servant'].astype(int)
house['security'] = house['security'].astype(int)
house['gas_pipeline'] = house['gas_pipeline'].astype(int)
house['park'] = house['park'].astype(int)
house['rain_water_harvest'] = house['rain_water_harvest'].astype(int)
house['sewage_treat_plant'] = house['sewage_treat_plant'].astype(int)
house['house_keeping'] = house['house_keeping'].astype(int)
```
